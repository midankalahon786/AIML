```python
# Forward propagation function
def sigmoid(z):
    """Return the sigmoid of z"""
    return 1 / (1 + np.exp(-z))
```

```python
# Prediction function for Neural Network
def predict(Theta1, Theta2, X):
    """Predict the label of an input given a trained neural network"""
    m = X.shape[0]
    
    # Add bias unit to input layer
    X = np.hstack((np.ones((m, 1)), X))
    
    # Forward propagation
    a1 = sigmoid(X @ Theta1.T)  # Hidden layer
    a1 = np.hstack((np.ones((m, 1)), a1))  # Add bias unit
    a2 = sigmoid(a1 @ Theta2.T)  # Output layer
    
    return np.argmax(a2, axis=1) + 1  # Return predicted labels (1-based index)
```

```python
# Load Thetas (Make sure Theta1 and Theta2 are defined)
pred2 = predict(Theta1, Theta2, X)
print("Training Set Accuracy:", np.mean(pred2[:, np.newaxis] == y) * 100, "%")
```

```python
# Cost Function for Neural Network
def nnCostFunction(nn_params, input_layer_size, hidden_layer_size, num_labels, X, y, Lambda):
    """Compute cost and gradient for a neural network"""
    
    # Unroll Theta1 and Theta2 from nn_params
    Theta1 = nn_params[:((input_layer_size + 1) * hidden_layer_size)].reshape(hidden_layer_size, input_layer_size + 1)
    Theta2 = nn_params[((input_layer_size + 1) * hidden_layer_size):].reshape(num_labels, hidden_layer_size + 1)

    m = X.shape[0]
    
    # Add bias unit to input layer
    X = np.hstack((np.ones((m, 1)), X))
    
    # Convert y to one-hot encoding
    y_onehot = np.zeros((m, num_labels))
    for i in range(num_labels):
        y_onehot[:, i] = (y == (i + 1)).flatten()
    
    # Forward propagation
    a1 = sigmoid(X @ Theta1.T)
    a1 = np.hstack((np.ones((m, 1)), a1))  # Add bias unit
    a2 = sigmoid(a1 @ Theta2.T)  # Output layer

    # Compute cost
    J = np.sum(-y_onehot * np.log(a2) - (1 - y_onehot) * np.log(1 - a2)) / m
    reg_J = J + (Lambda / (2 * m)) * (np.sum(Theta1[:, 1:] ** 2) + np.sum(Theta2[:, 1:] ** 2))

    # Backpropagation
    Theta1_grad = np.zeros(Theta1.shape)
    Theta2_grad = np.zeros(Theta2.shape)

    for i in range(m):
        xi = X[i, :]  # 1 x 401
        a1i = a1[i, :]  # 1 x 26
        a2i = a2[i, :]  # 1 x 10
        d2 = a2i - y_onehot[i, :]
        d1 = (Theta2.T @ d2.T) * sigmoid(np.hstack((1, xi @ Theta1.T)))  # Adding bias unit
        
        Theta1_grad += np.outer(d1[1:], xi)  # Removing bias unit effect
        Theta2_grad += np.outer(d2, a1i)

    # Average gradients
    Theta1_grad /= m
    Theta2_grad /= m

    # Regularization for gradients
    Theta1_grad[:, 1:] += (Lambda / m) * Theta1[:, 1:]
    Theta2_grad[:, 1:] += (Lambda / m) * Theta2[:, 1:]

    return reg_J, Theta1_grad, Theta2_grad
```

```python
def sigmoidGradient(z):
    """Computes the gradient of the sigmoid function"""
    sigmoid = 1 / (1 + np.exp(-z))
    return sigmoid * (1 - sigmoid)  # Fixed multiplication issue
```

```python
# Neural Network Hyperparameters
input_layer_size = 400
hidden_layer_size = 25
num_labels = 10

# Flatten and concatenate Thetas
nn_params = np.append(Theta1.flatten(), Theta2.flatten())  # Fixed typo

# Compute cost (non-regularized and regularized)
J, reg_J = nnCostFunction(nn_params, input_layer_size, hidden_layer_size, num_labels, X, y, 1)[::3]  # Fixed slicing

# Print cost
print(f"Cost at parameters (non-regularized): {J}")
print(f"Cost at parameters (Regularized): {reg_J}")
```

```python
def randInitializeWeights(L_in, L_out):
    """Randomly initializes the weights of a layer with L_in incoming connections and L_out outgoing connections."""
    epsilon = np.sqrt(6) / np.sqrt(L_in + L_out)  # Fixed epsilon calculation
    W = np.random.rand(L_out, L_in + 1) * (2 * epsilon) - epsilon  # Fixed scaling
    return W
```

```python
# Initialize weights for Theta1 and Theta2
initial_Theta1 = randInitializeWeights(input_layer_size, hidden_layer_size)  # Fixed function name
initial_Theta2 = randInitializeWeights(hidden_layer_size, num_labels)  # Fixed function name

# Flatten and concatenate the parameters
initial_nn_params = np.append(initial_Theta1.flatten(), initial_Theta2.flatten())
```

```python
# Compute cost with lambda = 3
debug_J = nnCostFunction(nn_params, input_layer_size, hidden_layer_size, num_labels, X, y, 3)

# Print cost
print("Cost at (fixed) debugging parameters (w/ lambda = 3):", debug_J[0])  # Fixed index
```

```python
def gradientDescentnn(X, y, initial_nn_params, alpha, num_iters, Lambda, input_layer_size, hidden_layer_size, num_labels):
    """
    Performs gradient descent to learn optimal neural network parameters.

    Parameters:
    X : ndarray
        Training data.
    y : ndarray
        Training labels.
    initial_nn_params : ndarray
        Flattened initial weight parameters.
    alpha : float
        Learning rate.
    num_iters : int
        Number of iterations.
    Lambda : float
        Regularization parameter.
    input_layer_size : int
        Number of input layer units.
    hidden_layer_size : int
        Number of hidden layer units.
    num_labels : int
        Number of output layer units.

    Returns:
    final_theta : ndarray
        Optimized neural network parameters.
    cost_history : list
        Cost at each iteration.
    """

    # Store cost history
    cost_history = []

    # Define cost function wrapper
    def costFunction(p):
        J, grad1, grad2, *_ = nnCostFunction(p, input_layer_size, hidden_layer_size, num_labels, X, y, Lambda)
        cost_history.append(J)
        return J, np.append(grad1.flatten(), grad2.flatten())

    # Use scipy.optimize.fmin_cg for optimization
    result = opt.fmin_cg(f=costFunction, x0=initial_nn_params, fprime=lambda p: costFunction(p)[1], maxiter=num_iters)

    return result, cost_history
```

```python
# Extract Theta1 and Theta2 from the flattened initial_nn_params
Theta1 = initial_nn_params[:(input_layer_size + 1) * hidden_layer_size] \
            .reshape(hidden_layer_size, input_layer_size + 1)

Theta2 = initial_nn_params[(input_layer_size + 1) * hidden_layer_size:] \
            .reshape(num_labels, hidden_layer_size + 1)
```

```python
m = len(y)
J_history = []
```

```python
for i in range(num_iters):
    # Compute cost and gradients
    cost, grad1, grad2 = nnCostFunction(
        np.append(Theta1.flatten(), Theta2.flatten()),
        input_layer_size, hidden_layer_size, num_labels, X, y, Lambda
    )[0:3]  # Extract cost and gradients correctly

    # Gradient update
    Theta1 -= alpha * grad1
    Theta2 -= alpha * grad2

    # Store cost history
    J_history.append(cost)

# Final flattened parameters
nn_paramsFinal = np.append(Theta1.flatten(), Theta2.flatten())

# Return the final parameters and cost history **outside** the loop
return nn_paramsFinal, J_history
```

```python
# Call the gradient descent function
nnTheta, nnJ_history = gradientDescentnn(
    X, y, initial_nn_params, alpha=0.8, num_iters=800, Lambda=1,
    input_layer_size=input_layer_size, hidden_layer_size=hidden_layer_size,
    num_labels=num_labels
)

# Correct reshaping for Theta1 and Theta2
Theta1 = nnTheta[:((input_layer_size + 1) * hidden_layer_size)] \
            .reshape(hidden_layer_size, input_layer_size + 1)

Theta2 = nnTheta[((input_layer_size + 1) * hidden_layer_size):] \
            .reshape(num_labels, hidden_layer_size + 1)
```

```python
pred3 = predict(Theta1, Theta2, X)

# Correct accuracy calculation
accuracy = np.mean(pred3[:, np.newaxis] == y) * 100

print(f"Training Set Accuracy: {accuracy:.2f}%")
```
